## Load Data
We use PIL library to load our images. Here we are creating our array where our input features are the mean colours and output features are the rotations along the x axis.

```python
!pip install -U scipy==1.2.0
```

```python
df = pd.read_csv(TRAINING_LABELS_PATH)
df = df.sort_values(by = 'xRot')
print(float(df.loc[df['filename'] == '004354.jpg', 'xRot']))
```

```python
def extract_features(image_path):
    return cv2.resize(cv2.imread(image_path,1), (64,64)).flatten()


def batch_extractor(images_path, pickled_db_path="features.pck"):
    files = [os.path.join(images_path, p) for p in sorted(os.listdir(images_path))]

    result = {}
    for f in files:
        print('Extracting features from image',f)
        name = f.split('/')[-1].lower()
        result[name] = extract_features(f)
    
    return result


class Matcher(object):

    def __init__(self, data):
        self.data = data
        self.names = []
        self.matrix = []
        for k, v in self.data.items():
            self.names.append(k)
            self.matrix.append(v)
        self.matrix = np.array(self.matrix)
        self.names = np.array(self.names)

    def cos_cdist(self, vector):
        # getting cosine distance between search image and images database
        v = vector.reshape(1, -1)
        return scipy.spatial.distance.cdist(self.matrix, v, 'cosine').reshape(-1)

    def match(self, image_path, topn=1):
        features = extract_features(image_path)
        img_distances = self.cos_cdist(features)
        # getting top 5 records
        nearest_ids = np.argsort(img_distances)[:topn].tolist()
        nearest_img_paths = self.names[nearest_ids].tolist()

        return nearest_img_paths, img_distances[nearest_ids].tolist()

def show_img(path):
    img = cv2.imread(path, 1)
    plt.imshow(img)
    plt.show()


images_path = TRAINING_IMAGES_FOLDER
files = [os.path.join(TEST_IMAGES_FOLDER, p) for p in sorted(os.listdir(TEST_IMAGES_FOLDER))]

res = batch_extractor(images_path)
```

We will take two nearest neighbours. For good matches, the rotation will be same as that of the first nearest neighbour. For bad matches, average of the two nearest neighbours will be taken

```python
ma = Matcher(res)

ans = []
ans2 = []

matches = []
cnt = 0
for i in tqdm.tqdm(range(len(files))):
    s = files[i]
    names, match = ma.match(s, topn=2)
    val = float(df.loc[df['filename'] == names[0], 'xRot'])
    val2 = float(df.loc[df['filename'] == names[1], 'xRot'])
    nam = s.split('/')[-1]
    ans.append((val ,nam))
    ans2.append((val2, nam))
    # print(i, (val ,nam))
    matches.append(match)
```

```python
## checking the bad matches
## Threshold for bad matches is set to 0.0005
cnt = 0
for i in range(len(matches)):
  if matches[i][0] > 0.0005:
    cnt += 1
    print(cnt, matches[i][0], matches[i][1])    
```

```python
## saving the predictions in an array to be later saved in a csv file

preds = []
names = []
for i in range(len(ans)):
  if matches[i][0] > 0.0005:
    preds.append((ans[i][0]+ans2[i][0])/2)
  else:
    preds.append(ans[i][0])
  names.append(ans[i][1])
print(preds)
print(names)
```

```python
test_df = pd.DataFrame(preds, columns=['xRot'])
test_df["filename"] = names
```

## Save the prediction to csv

```python
test_df.to_csv('submission.csv', index=False)
```
